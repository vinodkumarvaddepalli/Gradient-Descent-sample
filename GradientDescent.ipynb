###  Gradient Descent

```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
m=0
c=0
n=len(x)
iterations=1000
learning_rate=0.08
for i in range(iterations):
    y_pred=m*x+c
    cost=(1/n)*sum([val**2 for val in (y-y_pred)])
    md=-(2/n)*sum(x*(y-y_pred))
    cd=-(2/n)*sum(y-y_pred)
    m=m-learning_rate*md
    c=c-learning_rate*cd
    print("m value:{} and c value:{} and cost{} and iteartion {}".format(m,c,cost,i))
```
